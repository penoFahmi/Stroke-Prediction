# <a id='intro' href=#1> Background </a>

Stroke, atau yang dikenal sebagai penyakit stroke, adalah kondisi medis kritis yang terjadi akibat penyumbatan atau penurunan pasokan darah ke otak, yang menyebabkan kematian sel-sel saraf. Secara global, stroke memengaruhi lebih dari 12,2 juta kasus setiap tahunnya dan dapat terjadi pada individu dari semua kelompok demografi. Faktor risiko seperti hipertensi, penyakit jantung, usia, kebiasaan merokok, jenis kelamin, dan gaya hidup secara signifikan berkontribusi terhadap kerentanan seseorang terhadap stroke dengan memengaruhi kesehatan arteri dan aliran darah ke otak. Kondisi ini memberikan beban besar pada populasi dan sistem kesehatan, mengakibatkan angka kematian, kecacatan jangka panjang, dan peningkatan pengeluaran untuk layanan kesehatan. Dampak finansial perawatan stroke sangat bervariasi, dengan biaya rata-rata global berkisar antara €610 hingga €220.822. Selain itu, stroke juga berdampak signifikan pada kualitas hidup pasien, seperti menyebabkan kecacatan fisik, gangguan kognitif, depresi, dan isolasi sosial.

Dalam proyek ini, kami akan memanfaatkan dataset yang mencakup informasi demografis dan terkait kesehatan untuk membangun model prediktif guna menilai kemungkinan stroke pada pasien. Parameter input yang digunakan meliputi jenis kelamin, usia, keberadaan berbagai penyakit, dan status merokok.

Meskipun telah terjadi kemajuan signifikan dalam pencegahan dan penanganan stroke, identifikasi dini individu yang berisiko tinggi tetap menjadi tantangan besar. Estimasi risiko stroke yang tidak akurat pada tingkat individu dapat menghambat intervensi pencegahan yang tepat waktu dan menyebabkan alokasi sumber daya kesehatan yang tidak optimal.

Oleh karena itu, pengembangan model prediktif yang mampu mengidentifikasi individu dengan risiko tinggi stroke secara dini sangatlah penting. Prediksi yang berhasil memiliki potensi untuk meningkatkan hasil bagi pasien, memfasilitasi langkah-langkah pencegahan yang terfokus, serta mengoptimalkan alokasi sumber daya kesehatan, sehingga mendukung upaya pencegahan stroke dan pengelolaan penyakit secara keseluruhan.

**Sumber Data**  
Data diperoleh dari **Kaggle.com** dengan judul *"Stroke Prediction Dataset"* yang diterbitkan pada tahun 2020. Dataset ini berisi **5110 entri** dan **12 kolom**. Dataset ini digunakan untuk memprediksi apakah seorang pasien berpotensi mengalami stroke berdasarkan parameter input seperti jenis kelamin, usia, riwayat penyakit tertentu, dan status merokok. Setiap baris dalam dataset memberikan informasi relevan tentang pasien.  

**Informasi Atribut**  
1. **id**: Identifikasi unik untuk setiap pasien.  
2. **gender**: Jenis kelamin pasien, dapat berupa *"Male"*, *"Female"*, atau *"Other"*.  
3. **age**: Usia pasien.  
4. **hypertension**:  
   - **0**: Pasien tidak memiliki hipertensi.  
   - **1**: Pasien memiliki hipertensi.  
5. **heart_disease**:  
   - **0**: Pasien tidak memiliki penyakit jantung.  
   - **1**: Pasien memiliki penyakit jantung.  
6. **ever_married**: Status pernikahan pasien, berupa *"No"* atau *"Yes"*.  
7. **work_type**: Jenis pekerjaan pasien:  
   - *"children"*: Anak-anak.  
   - *"Govt_job"*: Pekerjaan pemerintah.  
   - *"Never_worked"*: Tidak pernah bekerja.  
   - *"Private"*: Pekerjaan di sektor swasta.  
   - *"Self-employed"*: Wiraswasta.  
8. **Residence_type**: Jenis tempat tinggal pasien, berupa *"Rural"* (pedesaan) atau *"Urban"* (perkotaan).  
9. **avg_glucose_level**: Rata-rata kadar glukosa dalam darah pasien.  
10. **bmi**: *Body Mass Index* (Indeks Massa Tubuh) pasien.  
11. **smoking_status**: Status merokok pasien:  
    - *"formerly smoked"*: Pernah merokok.  
    - *"never smoked"*: Tidak pernah merokok.  
    - *"smokes"*: Sedang merokok.  
    - *"Unknown"*: Status merokok tidak diketahui.  
12. **stroke**:  
    - **1**: Pasien pernah mengalami stroke.  
    - **0**: Pasien tidak pernah mengalami stroke.  

Dataset ini menjadi dasar untuk mengembangkan model prediktif guna mengidentifikasi risiko stroke pada pasien berdasarkan berbagai faktor yang telah diidentifikasi.

```python
from scipy.stats import chi2_contingency

# Fungsi untuk menghitung Cramér's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]  # Menghitung nilai chi-squared
    n = confusion_matrix.sum().sum()  # Jumlah total elemen dalam matriks kebingungannya
    phi2 = chi2/n  # Menghitung nilai phi-squared
    r, k = confusion_matrix.shape  # Mendapatkan dimensi matriks kebingungannya (baris dan kolom)
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))  # Koreksi untuk phi2
    rcorr = r-((r-1)**2)/(n-1)  # Koreksi untuk jumlah baris
    kcorr = k-((k-1)**2)/(n-1)  # Koreksi untuk jumlah kolom
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))  # Mengembalikan akar kuadrat dari Cramér's V

# Daftar kolom kategorikal
cat_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

# Inisialisasi dictionary kosong untuk menyimpan hasil Cramér's V
cramers_v_results = {}

# Menghitung Cramér's V untuk setiap fitur kategorikal
for col in cat_cols:
    confusion_matrix = pd.crosstab(df[col], df['stroke'])  # Matriks kebingungan antara fitur dan stroke
    cramers_v_results[col] = cramers_v(confusion_matrix.values)  # Menyimpan hasil Cramér's V

# Mengonversi hasil Cramér's V menjadi DataFrame
cramers_v_df = pd.DataFrame(list(cramers_v_results.items()), columns=['Feature', 'Cramér\'s V'])
cramers_v_df = cramers_v_df.sort_values(by='Cramér\'s V', ascending=False)  # Mengurutkan berdasarkan Cramér's V

# Membuat grafik batang
plt.figure(figsize=(10, 6))
sns.barplot(x='Cramér\'s V', y='Feature', data=cramers_v_df, palette='coolwarm')  # Plot Cramér's V
plt.title('Korelasi antara Fitur Kategorikal', fontsize=16)
plt.xlabel('Cramér\'s V', fontsize=14)
plt.ylabel('Fitur', fontsize=14)
plt.show()

# Menampilkan hasil Cramér's V
print("Hasil Cramér's V:")
print(cramers_v_df)
```

> **Hasil Pengamatan:**
> - Di antara fitur kategorikal, status "Pernah Menikah" dan "Jenis Pekerjaan" memiliki asosiasi yang cukup kuat namun masih lemah dengan risiko stroke.
> - "Status Merokok" mengikuti dengan asosiasi yang lemah namun tetap ada.
> - "Jenis Tempat Tinggal" memiliki asosiasi yang minimal, dan "Jenis Kelamin" tidak menunjukkan asosiasi dengan risiko stroke berdasarkan data yang diberikan.

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23, stratify=y)
```

```python
# Pipeline untuk menangani nilai hilang pada fitur numerik
numeric_missing_features = ['age', 'bmi','avg_glucose_level']
numeric_missing_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))  # Mengimputasi nilai hilang dengan median
])

# Pipeline untuk fitur numerik dengan StandardScaler
numeric_features = ['age', 'bmi','avg_glucose_level']
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),  # Mengimputasi nilai hilang dengan median
    ('scaler', StandardScaler())  # Menstandarisasi fitur numerik
])

# Pipeline untuk fitur kategorikal yang sudah dienkode
categorical_enc_features = ['hypertension','heart_disease']
categorical_enc_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Mengimputasi nilai hilang dengan nilai terbanyak
])

# Pipeline untuk fitur kategorikal ordinal
categorical_features = ['gender', 'ever_married']
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Mengimputasi nilai hilang dengan nilai terbanyak
    ('bin_encoder', OrdinalEncoder())  # Mengencode fitur kategorikal ordinal
])

# Pipeline untuk fitur kategorikal yang dienkode dengan One-Hot Encoding
categorical_ohe_features = ['work_type', 'Residence_type','smoking_status']
categorical_ohe_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Mengimputasi nilai hilang dengan nilai terbanyak
    ('onehot_encoder', OneHotEncoder())  # Menggunakan One-Hot Encoding untuk fitur kategorikal
])

# Menggabungkan semua pipeline
preprocessor = ColumnTransformer([
    ('numeric_missing', numeric_missing_transformer, numeric_missing_features),  # Menangani nilai hilang pada fitur numerik
    ('numeric', numeric_transformer, numeric_features),  # Menangani fitur numerik dengan standardisasi
    ('categorical_enc', categorical_enc_transformer, categorical_enc_features),  # Menangani fitur kategorikal yang sudah dienkode
    ('categorical_bin', categorical_transformer, categorical_features),  # Menangani fitur kategorikal ordinal
    ('categorical_ohe', categorical_ohe_transformer, categorical_ohe_features)  # Menangani fitur kategorikal dengan One-Hot Encoding
],
remainder='passthrough',  # Fitur lainnya tidak diproses
verbose=True)  # Menampilkan proses yang sedang dilakukan
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# Membuat objek SelectKBest untuk memilih 10 fitur terbaik berdasarkan uji Chi-Squared
selector = SelectKBest(chi2, k=10)  # Memilih 10 fitur terbaik
```

```python
from imblearn.over_sampling import SMOTENC
from imblearn.combine import SMOTETomek

# Determine categorical feature indices for SMOTENC
categorical_features_indices = list(range(1, len(X_train_new.columns)))

# Balancing using SMOTENC
smote_nc = SMOTENC(categorical_features=categorical_features_indices, random_state=42)
X_train_resampled, y_train_resampled = smote_nc.fit_resample(X_train_new, y_train)
# Balancing using SMOTETomek
smote_tomek = SMOTETomek(random_state=42)
X_train_resampled, y_train_resampled = smote_tomek.fit_resample(X_train_new, y_train)
```

2. Jika kita mencari keseimbangan antara precision dan recall, atau jika precision untuk pasien non-stroke (kelas 0) sangat penting, maka Decision Tree bisa menjadi pilihan yang lebih baik.
3. Dalam proyek ini, kita menghadapi masalah ketidakseimbangan kelas dalam dataset, di mana kelas positif (pasien berisiko stroke) memiliki frekuensi lebih rendah dibandingkan kelas negatif. Untuk mengatasi masalah ini, berikut adalah strategi yang digunakan:
   - **Oversampling dengan SMOTENC**:
     - Kita menggunakan SMOTENC untuk melakukan oversampling pada data pelatihan. Metode ini menghasilkan contoh sintetis dari kelas minoritas sambil mempertimbangkan fitur kategori dan numerik. Dengan oversampling, kita meningkatkan frekuensi kelas minoritas sehingga model bisa lebih efektif dalam mempelajari pola dari kelas ini.
   - **Pemilihan Fitur**:
     - Setelah proses oversampling, kita melakukan pemilihan fitur untuk menemukan fitur yang paling berpengaruh untuk prediksi. Ini dilakukan dengan menggunakan `permutation_importance` yang menghitung pentingnya setiap fitur dalam model. Fitur dengan skor penting lebih besar atau sama dengan 0.001 dipilih untuk digunakan dalam model.
   - **Evaluasi Berdasarkan Recall**:
     - Dalam proses pelatihan model, kita mengoptimalkan model untuk mencapai recall yang lebih baik dengan menggunakan `scoring='recall'` di GridSearchCV atau RandomizedSearchCV. Recall adalah metrik yang penting dalam kasus ini karena kita ingin mengidentifikasi sebanyak mungkin pasien yang berisiko stroke, meskipun ada beberapa kasus yang salah diklasifikasikan sebagai berisiko tinggi.
   
   Dengan mengimplementasikan kombinasi oversampling, pemilihan fitur, dan evaluasi berdasarkan recall, kita bertujuan untuk meminimalkan dampak ketidakseimbangan kelas, memungkinkan model untuk lebih akurat dalam mengidentifikasi pasien yang berisiko stroke.
   
4. Berdasarkan analisis pentingnya fitur yang kami lakukan menggunakan `permutation_importance`, fitur yang paling berpengaruh untuk memprediksi risiko stroke dalam proyek ini adalah:
   - **Usia**: Usia memiliki dampak signifikan dalam prediksi stroke. Hal ini sejalan dengan pemahaman medis karena usia adalah faktor risiko yang sudah dikenal untuk stroke.
   - **Rata-rata Tingkat Glukosa**: Rata-rata tingkat glukosa juga memainkan peran penting dalam memprediksi risiko stroke. Tingkat glukosa yang tinggi dapat menyebabkan berbagai masalah kesehatan, termasuk penyakit kardiovaskular seperti stroke.
