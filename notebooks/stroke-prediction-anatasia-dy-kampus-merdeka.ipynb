# <a id='intro' href=#1> Background </a>

Stroke, atau yang dikenal sebagai penyakit stroke, adalah kondisi medis kritis yang terjadi akibat penyumbatan atau penurunan pasokan darah ke otak, yang menyebabkan kematian sel-sel saraf. Secara global, stroke memengaruhi lebih dari 12,2 juta kasus setiap tahunnya dan dapat terjadi pada individu dari semua kelompok demografi. Faktor risiko seperti hipertensi, penyakit jantung, usia, kebiasaan merokok, jenis kelamin, dan gaya hidup secara signifikan berkontribusi terhadap kerentanan seseorang terhadap stroke dengan memengaruhi kesehatan arteri dan aliran darah ke otak. Kondisi ini memberikan beban besar pada populasi dan sistem kesehatan, mengakibatkan angka kematian, kecacatan jangka panjang, dan peningkatan pengeluaran untuk layanan kesehatan. Dampak finansial perawatan stroke sangat bervariasi, dengan biaya rata-rata global berkisar antara €610 hingga €220.822. Selain itu, stroke juga berdampak signifikan pada kualitas hidup pasien, seperti menyebabkan kecacatan fisik, gangguan kognitif, depresi, dan isolasi sosial.

Dalam proyek ini, kami akan memanfaatkan dataset yang mencakup informasi demografis dan terkait kesehatan untuk membangun model prediktif guna menilai kemungkinan stroke pada pasien. Parameter input yang digunakan meliputi jenis kelamin, usia, keberadaan berbagai penyakit, dan status merokok.

Meskipun telah terjadi kemajuan signifikan dalam pencegahan dan penanganan stroke, identifikasi dini individu yang berisiko tinggi tetap menjadi tantangan besar. Estimasi risiko stroke yang tidak akurat pada tingkat individu dapat menghambat intervensi pencegahan yang tepat waktu dan menyebabkan alokasi sumber daya kesehatan yang tidak optimal.

Oleh karena itu, pengembangan model prediktif yang mampu mengidentifikasi individu dengan risiko tinggi stroke secara dini sangatlah penting. Prediksi yang berhasil memiliki potensi untuk meningkatkan hasil bagi pasien, memfasilitasi langkah-langkah pencegahan yang terfokus, serta mengoptimalkan alokasi sumber daya kesehatan, sehingga mendukung upaya pencegahan stroke dan pengelolaan penyakit secara keseluruhan.

# <a id='2' href=#1> Objectives </a>


In the discourse surrounding stroke prevention and management, each outlined objective holds significant importance; however, the early identification of high-risk individuals emerges as particularly pivotal. Here's why:

1. **Life-Preserving Potential:** Identifying individuals at elevated risk of stroke enables timely interventions and treatments, potentially mitigating the likelihood of stroke occurrence and ameliorating patient prognosis. Given stroke's acute nature and propensity for irreversible damage, prompt intervention stands as a cornerstone in averting adverse outcomes and preserving lives.

2. **Preventive Medicine Imperative:** The proactive identification of high-risk individuals underscores the essence of preventive medicine, emphasizing preemptive measures to forestall disease onset. This paradigmatic shift from reactive to proactive healthcare strategies aligns with broader public health principles, advocating for upstream interventions to curtail disease burden.

3. **Resource Allocation Optimization:** Early risk stratification facilitates targeted resource allocation, directing interventions and treatments toward individuals deemed most vulnerable. This judicious distribution of resources optimizes healthcare efficiency by maximizing impact while minimizing resource wastage—a crucial consideration in resource-constrained healthcare settings.

4. **Societal Burden Mitigation:** Stroke imposes a substantial societal burden, encompassing morbidity, mortality, and healthcare expenditures. By identifying high-risk individuals early and implementing preventive measures, the societal burden associated with stroke-related disability and healthcare costs can be significantly mitigated, fostering societal resilience and well-being.

```python
from scipy.stats import chi2_contingency

# Function to calculate Cramér's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))

# List of categorical columns
cat_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

# Initialize an empty dictionary to store Cramér's V results
cramers_v_results = {}

# Calculate Cramér's V for each categorical feature
for col in cat_cols:
    confusion_matrix = pd.crosstab(df[col], df['stroke'])
    cramers_v_results[col] = cramers_v(confusion_matrix.values)

# Convert cramers_v_results to a DataFrame
cramers_v_df = pd.DataFrame(list(cramers_v_results.items()), columns=['Feature', 'Cramér\'s V'])
cramers_v_df = cramers_v_df.sort_values(by='Cramér\'s V', ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Cramér\'s V', y='Feature', data=cramers_v_df, palette='coolwarm')
plt.title('Correlation Among Categorical Features', fontsize=16)
plt.xlabel('Cramér\'s V', fontsize=14)
plt.ylabel('Features', fontsize=14)
plt.show()

print("Cramér's V Results:")
print(cramers_v_df)
```

> **Obsertvation result:**
> - Among the categorical features, being "Ever Married" and "Work Type" have the strongest weak associations with stroke risk.
> - "Smoking Status" follows closely but still maintains a weak association.
> - "Residence Type" has a minimal association, and "Gender" shows no association with stroke risk based on the provided data.

# <a id='9' href=#1> Data Prepocessing </a>

```python
X= df.drop(['stroke','id'],axis=1)
y= df['stroke']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23, stratify=y)
```

```python
# Pipeline for missing values in numerical features
numeric_missing_features = ['age', 'bmi','avg_glucose_level']
numeric_missing_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])
# Pipeline for numerical features with StandardScaler
numeric_features = ['age', 'bmi','avg_glucose_level']
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pipeline for encoded categorical features
categorical_enc_features = ['hypertension','heart_disease']
categorical_enc_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# Pipeline for ordinal categorical features
categorical_features = ['gender', 'ever_married']
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('bin_encoder', OrdinalEncoder())
])

# Pipeline for one-hot encoded categorical features
categorical_ohe_features = ['work_type', 'Residence_type','smoking_status']
categorical_ohe_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot_encoder', OneHotEncoder())
])

# Combine all pipelines
preprocessor = ColumnTransformer([
    ('numeric_missing', numeric_missing_transformer, numeric_missing_features),
    ('numeric', numeric_transformer, numeric_features),
    ('categorical_enc', categorical_enc_transformer, categorical_enc_features),
    ('categorical_bin', categorical_transformer, categorical_features),
    ('categorical_ohe', categorical_ohe_transformer, categorical_ohe_features)
],
remainder='passthrough',
verbose=True)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

selector = SelectKBest(chi2, k=10)  # Select top 10 features
```

# <a id='11' href=#1> Conclusions </a>

1.  If the goal is to prioritize recall and correctly identify as many stroke patients as possible, Random Forest might be preferred due to its high recall score for class 1.
2.  If we're looking for a balance between precision and recall, or if the precision for non-stroke patients (class 0) is crucial, Decision Tree might be a better choice.
3. In this project, we face a class imbalance issue in the dataset, where the positive class (patients at risk of stroke) has a lower frequency compared to the negative class. To address this issue, here are the strategies we use:
    - **Oversampling with SMOTENC**:
        - We use SMOTENC to perform oversampling on the training data. This method generates synthetic examples from the minority class while considering both categorical and numerical features. With oversampling, we increase the frequency of the minority class so that the model can learn patterns from this class more effectively.
    - **Feature Selection**:
        - After the oversampling process, we perform feature selection to find the most influential features for prediction. This is done using `permutation_importance` which calculates the importance of each feature in the model. Features with an importance score greater than or equal to 0.001 are selected for use in the model.
    - **Evaluation Based on Recall**:
        - In the model training process, we optimize the model to achieve better recall by using `scoring='recall'` in GridSearchCV or RandomizedSearchCV. Recall is an important metric in this case because we want to identify as many patients at risk of stroke as possible, even if there are some cases misclassified as high risk.
        
   By implementing a combination of oversampling, feature selection, and evaluation based on recall, we aim to minimize the impact of class imbalance, allowing the model to more accurately identify patients at risk of stroke.
4. Based on the feature importance analysis we conducted using `permutation_importance`, the most influential features for predicting stroke risk in our project are:
    - **Age**: Age has a significant impact on stroke prediction. This aligns with medical understanding as age is a well-known risk factor for stroke.
    - **Average Glucose Level**: Average glucose level also plays a crucial role in predicting stroke risk. High glucose levels can lead to various health issues, including cardiovascular diseases like stroke.
    - **BMI (Body Mass Index)**: BMI is another important feature that affects stroke prediction. A higher BMI often indicates obesity, which is associated with an increased risk of stroke.
    These features were selected based on their importance scores, which were above or equal to 0.001 in our permutation importance analysis. By focusing on these influential features, our model can make more accurate predictions regarding stroke risk.
